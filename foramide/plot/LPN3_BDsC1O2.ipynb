```python
import matplotlib.pyplot as plt
```

```python
dihedrals = [10*ang for ang in range(19)]
ANG_H5_BD1 = []
ANG_H5_BD2 = []
E2_LPaBD1 = []
E2_LPaBD2 = []

for dihedral in dihedrals:
    
    with open(f'../nbo/dihedral_{dihedral:03d}/nbo.out','r') as file:
        lines = file.readlines()
    file.close()

    for i in range(len(lines)):
        if lines[i].startswith("    6. LP ( 1) N  3            13. BD*( 1) C  1- O  2"):
            print(lines[i].strip())
            ANG_H5_BD1.append(dihedral)
            E2_LPaBD1.append(float(lines[i].split()[13]))

        if lines[i].startswith("    6. LP ( 1) N  3            14. BD*( 2) C  1- O  2"):
            # ANG_H5.append(dihedral)
            ANG_H5_BD2.append(dihedral)
            E2_LPaBD2.append(float(lines[i].split()[13]))
```

```python
fig,ax = plt.subplots(figsize=(5,5))

ax.plot(ANG_H5_BD1,E2_LPaBD1,marker='h', markersize=4,color='crimson',    label='LP(1)N3 -> BD*(1)C1-O2')
ax.plot(ANG_H5_BD2,E2_LPaBD2,marker='h', markersize=4,color='darkblue',    label='LP(1)N3 -> BD*(2)C1-O2')

ax.set_xticks(ANG_H5_BD1)

ax.set_xlabel('O2=C1-N3-H5 Dihedral angle')
ax.set_ylabel('E(2), kcal/mol')

ax.grid(alpha=0.25)
ax.legend()

plt.savefig('LPN3_BDsC1O2.pdf')
plt.show()
```
